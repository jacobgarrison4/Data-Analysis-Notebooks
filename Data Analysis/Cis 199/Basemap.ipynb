```python
from mpl_toolkits.basemap import Basemap
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
plt.figure(figsize=(8,8))
m = Basemap(projection='ortho', resolution=None, lat_0 = 45.0, lon_0 = -120.0)
m.bluemarble();
```

```python
m = Basemap(width=12_000_000,
           height=9_000_000,
           projection='lcc',
           resolution=None,
           lat_1=45., lat_2=55,
           lat_0=50, lon_0=-107.)

m.drawlsmask(land_color='coral',
            ocean_color='aqua',
            lakes=True)

plt.show()
```

```python
fig2 = plt.figure(figsize=(8,8))
m = Basemap(projection='lcc',
           resolution=None,
           width=4E6, height=4E6,
           lat_0=51, lon_0=-1,)

x, y = m(0.0001, 51.482578)
plt.plot(x, y, 'ok', markersize=4)
plt.text(x, y, 'Greenwich')

x, y = m(6.6323, 46.5197)
plt.plot(x, y, 'ok', markersize=4)
plt.text(x, y, 'Lausanne')

m.etopo(scale=1, alpha=0.50)
m.drawmeridians([])
```
