```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = 'https://classes.cs.uoregon.edu/18F/cis199ids/data/' #names 1880-2017
file = 'names_1880_2017.csv'
url = path + file
names_df = pd.read_csv(url)
names_df.shape
```

This took about 5 seconds to load, and got me almost 2 million names

```python
found_df = names_df[(names_df['name'] == 'Jacob') & (names_df['gender'] == 'M')]
print('Jacob'.title(), 'M'.title())
found_df.shape
```

```python
found_df.head()
```

```python
sex = 'M'.title()
gender_name = 'Boys'
if sex == 'F':
    gender_name = 'Girls'
my_title = gender_name + ' named ' + 'Jacob'.title()
found_df.plot('year', #x value
             'number', #y value
             kind='line', 
             title=my_title, 
             color='red');
```
