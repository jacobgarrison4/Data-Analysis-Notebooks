```python
import pandas as pd
import numpy as np
```

```python
path = 'https://classes.cs.uoregon.edu/18F/cis199ids/data/us_state_code.csv'
states_df = pd.read_csv(path, index_col='code')
```

```python
states_df.head()
```

```python
states_df.describe()
```

```python
pd.to_numeric(states_df['city_pop'], errors='coerce')
states_df['city_pop'].fillna(0.000, inplace=True)

states_df.head(10)
```

```python
states_df.sample(frac=.1)
```

```python
temp = states_df['city_pop'].values
for item in temp:
    if type(item) != np.float64:
        print(item, type(item))
```

```python
states_df.pivot_table(index=['locale']) #note: these are averages
```

```python
states_df.pivot_table(index=['locale', 'year'])
```

```python
states_df.pivot_table(values=['pop_millions'],   #values are numbers to use
              aggfunc=[np.median],        #aggregation function names(s)
              index=['locale'])
```

```python
states_df.pivot_table(values=['pop_millions'],   #values are numbers to use
              aggfunc=[np.median, np.mean, np.min, np.max],        #aggregation function names(s)
              index=['locale'])
```

```python
states_df.pivot_table(values=['pop_millions'], #calculate using this
              aggfunc=[np.median], #get median value
                columns=['year'], #across
              index=['locale'],#down
                     fill_value=0)  #fill 'NaN'
```
