The pivot table that gives the best insight into the size of tip would be the last one used, which has the size of the party and the total bill plus the tip. Clearly, the bigger the size, the bigger the total bill, and the bigger the tip.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = pd.read_csv("http://www.cs.uoregon.edu/classes/18F/cis199ids/data/tips.csv")
cali = pd.read_csv("http://www.cs.uoregon.edu/classes/18F/cis199ids/data/california_cities.csv")
```

```python
pd.pivot_table(tips, index=['day'])
```

```python
pd.pivot_table(tips, index=['day', 'size'])
```

```python
pd.pivot_table(tips, index=['day', 'size', 'sex'])
```

```python
pd.pivot_table(tips, index=['day', 'size', 'sex'], values=['tip', 'total_bill'])
```

```python
pd.pivot_table(tips, index=['day', 'size', 'sex'], values=['tip', 'total_bill'], aggfunc=[np.median, np.sum])
```

```python
pd.pivot_table(tips, index=['day'], values=['tip', 'total_bill'], aggfunc=[np.median, np.sum])
```

```python
pd.pivot_table(tips, index=['smoker'], values=['tip', 'total_bill'], aggfunc=[np.median, np.sum])
```

```python
pd.pivot_table(tips, index=['sex'], values=['tip', 'total_bill'], aggfunc=[np.median, np.sum])
```

```python
pd.pivot_table(tips, index=['size'], values=['tip', 'total_bill'], aggfunc=[np.median, np.sum])
```

```python
pd.pivot_table(tips, index=['size'], values=['tip', 'total_bill'], aggfunc=[np.average])
#This is the most useful table for a waiter who wants more in tips because it shows the average tip amount compared to the
#total bill per party size
```
