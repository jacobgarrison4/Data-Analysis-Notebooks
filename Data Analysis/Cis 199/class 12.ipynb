```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pd.options.display.max_rows = 200
path = 'https://classes.cs.uoregon.edu/18F/cis199ids/data/' #names 1880-2017
file = 'names_1880_2017.csv'
url = path + file
names_df = pd.read_csv(url)
names_df.shape
```

```python
found_df = names_df[(names_df['name'] == 'Melissa') & (names_df['gender'] == 'F')]
print('Melissa'.title(), 'F'.title())
found_df.shape
```

```python
found_df.head()
```

```python
sex = 'f'.title()
gender_name = 'Girls'
if sex == 'M':
    gender_name = 'Boys'
my_title = gender_name + ' named ' + 'Melissa'.title()
found_df.plot('year', #x value
             'number', #y value
             kind='line', 
             title=my_title, 
             color='red');
```
